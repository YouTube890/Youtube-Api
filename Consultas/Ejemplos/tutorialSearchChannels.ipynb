### Searching Oldest Youtube Videos

```python
from datetime import datetime

datetime(year=2005, month=1, day=1)
```

```python
datetime(year=2005, month=1, day=1).isoformat()
```

```python
datetime(year=2005, month=1, day=1).strftime('%Y-%m-%dT%H:%SZ')
```

```python
# La api pide adicionarle la Z al final y el formato RFC 3339
start_time=datetime(year=2005, month=1, day=1).isoformat()+'Z'
end_time = datetime(year=2006, month=1, day=1).isoformat()+'Z'
```

```python
req = youtube.search().list(
    q='python programming',
    part='snippet',
    type='video',
    maxResults=50,
    publishedAfter=start_time,
    publishedBefore=end_time)
res = req.execute()
res
```

```python
for item in res['items']:
    print(item['snippet']['title'],"..." ,item['snippet']['publishedAt'],"...",item['id']['videoId'])
```

#### Funci√≥n que da la lista de los videos en un canal



```python
from apiclient.discovery import build

api_key = "AIzaSyDqq-BEw8S5_mJq60d5QGZcBv13Gyu6qiw"
youtube = build('youtube','v3',developerKey=api_key)

channel_id = 'UNsedebogota'

def get_channel_videos(channel_id):
    res =  youtube.channels().list(forUsername=channel_id,
                                  part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    videos = []
    next_page_token = None
    
    while 1:
        res = youtube.playlistItems().list(playlistId = playlist_id,
                                          part='snippet',
                                          maxResults=50,
                                          pageToken=next_page_token).execute()
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None:
            break
        
    return videos

videos = get_channel_videos(channel_id)

for video in videos:
    print(video['snippet']['title'])
```
